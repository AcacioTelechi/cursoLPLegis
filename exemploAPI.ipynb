```python
import requests
import pandas as pd
```

```python
palavras_chave = ['meio ambiente']

ano = '2021'

url = f"https://dadosabertos.camara.leg.br/api/v2/proposicoes?keywords={palavras_chave}&ano={ano}&itens=50"

resposta = requests.get(url)

resposta_json = resposta.json()
```

```python
dados = resposta_json['dados']
```

```python
df = pd.DataFrame(dados)
```

```python
df.to_excel('proposicoes.xlsx')
```
